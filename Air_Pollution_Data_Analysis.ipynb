# Data Anaylsis

### Summarizing Data

```python
air_pollution
```

```python
#filter by specific states
states_to_filter = ['Kentucky', 'Tennessee', 'Ohio']
air_pollution_filtered = air_pollution[air_pollution['State'].isin(states_to_filter)]

#KY, OH, TN group by years
air_pollution_avg = air_pollution_filtered.groupby(['State', 'Year']).agg({'O3 Mean': 'mean', 'CO Mean': 'mean', 'SO2 Mean': 'mean', 'NO2 Mean': 'mean'}).reset_index()
print(air_pollution_avg)
```

```python
#Avg Data for all states
all_averaged = air_pollution.groupby(['State', 'Year']).agg({'O3 Mean': 'mean', 'CO Mean': 'mean', 'SO2 Mean': 'mean', 'NO2 Mean': 'mean'}).reset_index()
print(all_averaged)

#Group by year 2010
data_2010 = all_averaged[all_averaged['Year'] == 2010]
print(data_2010)

#Group by year 2020
data_2020 = all_averaged[all_averaged['Year'] == 2020]
print(data_2020)

#Group by year 2010 and 2020
data_2013_2023 = all_averaged[all_averaged['Year'].isin([2013, 2023])]
print(data_2013_2023)
```

```python
#Information on O3 Mean, CO Mean, SO2 Mean, and NO2 Mean

desc_specific = air_pollution[['O3 Mean', 'CO Mean', 'SO2 Mean', 'NO2 Mean',]].describe()
print(desc_specific)
```

### Plots

```python
#NO2 Mean Line Plot over the years for KY, TN, OH

#filter data for each state 
df_KY = air_pollution_avg[air_pollution_avg['State'] == 'Kentucky']
df_OH = air_pollution_avg[air_pollution_avg['State'] == 'Ohio']
df_TN = air_pollution_avg[air_pollution_avg['State'] == 'Tennessee']

#Create line plot of NO2 Mean for each state
plt.plot(df_KY['Year'], df_KY['NO2 Mean'], marker='o', linestyle='-', color='b', label='KY')
plt.plot(df_OH['Year'], df_OH['NO2 Mean'], marker='o', linestyle='-', color='r', label='OH')
plt.plot(df_TN['Year'], df_TN['NO2 Mean'], marker='o', linestyle='-', color='g', label='TN')

#Add labels and title
plt.title('NO2 Mean Over Years for KY, TN, and OH')
plt.xlabel('Year')
plt.ylabel('NO2 Mean')


#Add legend
plt.legend()

#Show the plot
plt.show()
```

# Summary of Data

#### CO
There appears to be a general downward trend across the states with only a couple of states going up over the ten year period. When looking at KY, OH, and TN specifically over the entire dataset, there appears to be some fluctuation in the trends. 

#### NO2
There appears to be a general downward trend across the states with only a couple going up over the ten year period. When looking at KY, OH, and TN specifically over the entire dataset, there appears to be a general downward trend except with TN which has stayed about the same, and started to rise. 

#### O3
There appears to be a general upward trend across the states with only a couple going down over the ten year period. When looking at KY, OH, and TN specifically over the entire dataset, there appears to be some major fluctuations, however, in recent years, there appears to be an upward trend. 

#### Summarization and Future Recommendations
There is a general downward trend for NO2 and CO and an upward trend for O3. Further analysis could include comparing O3, NO2, and CO levels to industrial changes over time as well as population changes over time.
