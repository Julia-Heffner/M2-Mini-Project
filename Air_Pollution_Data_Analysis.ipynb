```python
#load the data 
air_pollution = pd.read_csv('pollution_2000_2023.csv')
```

# Cleaning the Data

```python
#Missing values
air_cleaned = air_pollution.isnull().any().any()
print(air_cleaned)
```

```python
air_cleaned = air_pollution.isnull().any(axis=1)
print(air_cleaned)
```

```python
columns = list(air_pollution.columns)
print(columns)
```

```python
air_pollution.drop(['Unnamed: 0', 'Address', 'County', 'City', 'O3 1st Max Value', 'O3 1st Max Hour', 'O3 AQI', 'CO 1st Max Value', 'CO 1st Max Hour','CO AQI', 'SO2 1st Max Value', 'SO2 1st Max Hour','SO2 AQI', 'NO2 1st Max Value', 'NO2 1st Max Hour', 'NO2 AQI'], axis=1, inplace=True)
```

```python
# Separating Month and Year from Date Column
air_pollution['Date'] = pd.to_datetime(air_pollution['Date'])
air_pollution['Year'] = air_pollution['Date'].dt.year
air_pollution['Month'] = air_pollution['Date'].dt.month
```

```python
columns = list(air_pollution.columns)
print(columns)
```

# Data Anaylsis

### Summarizing Data

```python
air_pollution
```

```python
#filter by specific states
states_to_filter = ['Kentucky', 'Tennessee', 'Ohio']
air_pollution_filtered = air_pollution[air_pollution['State'].isin(states_to_filter)]
```

```python
#group by years
air_pollution_avg = air_pollution_filtered.groupby(['State', 'Year']).agg({'O3 Mean': 'mean', 'CO Mean': 'mean', 'SO2 Mean': 'mean', 'NO2 Mean': 'mean'}).reset_index()
print(air_pollution_avg)
```

```python
#Information on O3 Mean, CO Mean, SO2 Mean, and NO2

desc_specific = air_pollution[['O3 Mean', 'CO Mean', 'SO2 Mean', 'NO2 Mean',]].describe()
print(desc_specific)
```

### Plots

```python
#NO2 Mean Line Plot over the years for KY, TN, OH

#filter data for each state 
df_KY = air_pollution_avg[air_pollution_avg['State'] == 'Kentucky']
df_OH = air_pollution_avg[air_pollution_avg['State'] == 'Ohio']
df_TN = air_pollution_avg[air_pollution_avg['State'] == 'Tennessee']

#Create line plot of NO2 Mean for each state
plt.plot(df_KY['Year'], df_KY['NO2 Mean'], marker='o', linestyle='-', color='b', label='KY')
plt.plot(df_OH['Year'], df_OH['NO2 Mean'], marker='o', linestyle='--', color='b', label='OH')
plt.plot(df_TN['Year'], df_TN['NO2 Mean'], marker='o', linestyle='-.', color='b', label='TN')

#Add labels and title
plt.title('NO2 Mean Over Years for KY, TN, and OH')
plt.xlabel('Year')
plt.ylabel('NO2 Mean')

#Add legend
plt.legend()

#Show the plot
plt.show()
```
